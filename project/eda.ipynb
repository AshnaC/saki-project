```python
prefixes = get_file_prefixes()
print(prefixes)
for prefix in prefixes:
    cols = df.columns.str.startswith(prefix)
    cols = df.columns[cols]
    filtered_df = df[cols].fillna(value=np.nan)
    formatted_cols = [i.replace(prefix+'--', '') for i in cols]
    filtered_df.astype(float).dropna().plot(figsize=(10, 5))
    plt.gca().legend(formatted_cols)
    plt.title(prefix)
```
